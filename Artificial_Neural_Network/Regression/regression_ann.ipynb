```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
BHNames= ['crim','zn','indus','chas','nox','rm',
         'age','dis','rad','tax','ptratio','black','lstat','medv']
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
data = pd.read_csv(url, delim_whitespace=True,names=BHNames)
data.head()
```

```python
X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values
X.shape
```

```python
data.info()
```

```python
summary = data.describe()
summary = summary.transpose()
summary
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
Datascaled = ms.fit_transform(data)
Datascaled = pd.DataFrame(Datascaled,columns=BHNames)
Datascaled.shape
```

```python
summary = Datascaled.describe()
summary = summary.transpose()
summary
```

```python
box_plot = Datascaled.boxplot(column=BHNames,figsize=(15,10))
plt.show()
```

```python
CorData = Datascaled.corr(method='pearson')
print(CorData)
```

```python
import seaborn as sns
# sns.heatmap(CorData)
mask = np.zeros_like(CorData)  
mask[np.triu_indices_from(mask)] = True  
with sns.axes_style("white"):  
    ax = sns.heatmap(CorData, mask=mask, vmax=.3, square=True) 
```

```python
X = Datascaled.drop('medv',axis=1)
Y = Datascaled['medv']
print(X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15,random_state=42)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model = Sequential()
model.add(Dense(20,init='uniform',activation='relu',input_dim=13))
model.add(Dense(10,init='uniform',activation='relu'))
model.add(Dense(1,activation='linear'))
model.compile('Adam','mean_squared_error',['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=1000,verbose=1)
```

```python
Y_predKM = model.predict(X_test)
```

```python
acc,loss = model.evaluate(X_test,Y_test,verbose=0)
```

```python
acc,loss
```

```python
#Multiple Linear Regression
from sklearn.linear_model import LinearRegression

LModel = LinearRegression()
LModel.fit(X_train, Y_train)

Y_predLM = LModel.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, Y_predLM)
print('Linear Regression Model Mean Square Error')
print(mse)
```
