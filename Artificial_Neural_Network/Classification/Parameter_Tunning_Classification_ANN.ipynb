```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
del data['Unnamed: 32']
```

```python
X = data.iloc[:,2:].values
Y = data.iloc[:,1].values
print([X.shape, Y.shape])
```

```python
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
Y = lc.fit_transform(Y)
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV 
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(16,init = 'uniform', activation='relu',input_dim=30))
    classifier.add(Dense(16,init = 'uniform',activation='relu'))
    classifier.add(Dense(1,activation='sigmoid'))
    classifier.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
clf = KerasClassifier(build_fn=build_classifier)
parameters = {'batch_size': [10, 32],'epochs': [100, 150],'optimizer': ['adam', 'rmsprop']}
grid_search = GridSearchCV(estimator=clf,param_grid=parameters,cv=10,scoring='accuracy')
grid_search = grid_search.fit(X_train,Y_train)
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_
```
